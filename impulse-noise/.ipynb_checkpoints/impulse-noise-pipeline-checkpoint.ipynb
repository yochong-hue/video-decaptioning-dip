```python
import os
import cv2

def extract_debug_frames_from_videos(results_dict, video_keys, frame_idx, out_dir="_final_debug_frames"):
    """
    Extract frame `frame_idx` from selected videos listed in `video_keys`,
    where videos come from the dictionary returned by full_process().

    Parameters
    ----------
    results_dict : dict
        Output from full_process(), containing filenames of generated videos.
    video_keys : list of str
        Keys in results_dict to extract frames from.
        Example: ["original_video", "noised_video", "denoised_video",
                  "original_mask", "noised_mask", "denoised_mask"]
    frame_idx : int
        Frame index to extract (0-indexed).
    out_dir : str
        Directory where frames will be saved.
    """
    os.makedirs(out_dir, exist_ok=True)

    for key in video_keys:
        if key not in results_dict:
            print(f"[WARN] Key '{key}' not found in results_dict. Skipping.")
            continue

        video_path = results_dict[key]
        video_path = os.path.expanduser(video_path)

        cap = cv2.VideoCapture(video_path)
        if not cap.isOpened():
            print(f"[ERROR] Cannot open video: {video_path}")
            continue

        total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        if frame_idx >= total:
            print(f"[WARN] Video '{key}' has only {total} frames. Requested {frame_idx}. Skipping.")
            cap.release()
            continue

        # Seek to frame
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_idx)
        ret, frame = cap.read()
        cap.release()

        if not ret:
            print(f"[ERROR] Failed to read frame {frame_idx} from {video_path}")
            continue

        # Convert to RGB for consistency
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        out_name = f"{key}_frame_{frame_idx}.png"
        out_path = os.path.join(out_dir, out_name)
        cv2.imwrite(out_path, cv2.cvtColor(frame_rgb, cv2.COLOR_RGB2BGR))

        print(f"[SAVED] {out_path}")

    print(f"[DONE] Extracted frames → {out_dir}")
```

# Decaption Network

```python
import os
import sys
import cv2
import imageio
import numpy as np
import torch
from PIL import Image

## The following two lines has been added in the video_pipeline.py
# SHARED_ROOT = os.path.expanduser("~/teams/imagenie")
# sys.path.append(f"{SHARED_ROOT}/teams/imagenie/Video_Decaptioning/")

from video_decaption.dataset.dataset import videopadding
from video_decaption.model.network import generator

```

```python


# =====================
# Configuration
# =====================
BASE = "./"

FRAME_ROOT  = os.path.join(BASE, "frame_{}")
MASK_ROOT   = os.path.join(BASE, "mask_{}")
OUTPUT_ROOT = os.path.join(BASE, "results/recovered_{}")

METHODS = {
    "raw": ("raw", "orig"),
}

model_G_path = "teams/imagenie/Video_Decaptioning/video_decaption/checkpoint/netG.pth"

T = 7
s = 3
FPS = 25

os.environ["CUDA_VISIBLE_DEVICES"] = "0"
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# =====================
# Load generator
# =====================
net_G = generator()
net_G.load_state_dict(torch.load(model_G_path, map_location=DEVICE))
net_G = net_G.to(DEVICE)
net_G.eval()
```

```python

# =====================
# Core function
# =====================
def recover_video(original_video, mask_video, out_video, out_dir):

    os.makedirs(out_dir, exist_ok=True)

    cap_vid = cv2.VideoCapture(original_video)
    cap_msk = cv2.VideoCapture(mask_video)

    assert cap_vid.isOpened(), "Cannot open original video"
    assert cap_msk.isOpened(), "Cannot open mask video"

    frames = []
    masks  = []

    # =====================
    # Read videos
    # =====================
    while True:
        ret_v, frame = cap_vid.read()
        ret_m, mask  = cap_msk.read()

        if not ret_v or not ret_m:
            break

        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB) / 255.0

        if mask.ndim == 3:
            mask = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)
        mask = (mask > 127).astype(np.float32)

        frames.append(frame)
        masks.append(mask)

    cap_vid.release()
    cap_msk.release()

    assert len(frames) > 0, "No frames read"
    assert len(frames) == len(masks), "Video length mismatch"

    n_frames, h, w, _ = np.array(frames).shape

    frames_np = np.stack(frames).astype(np.float32)
    masks_np  = np.stack(masks).astype(np.float32)[:, None, :, :]

    # =====================
    # Convert to tensors
    # =====================
    frames_t = torch.from_numpy(frames_np).permute(0, 3, 1, 2)
    frames_t = (frames_t - 0.5) / 0.5
    frames_t = frames_t.unsqueeze(0).to(DEVICE)

    masks_t = torch.from_numpy(masks_np).unsqueeze(0).to(DEVICE)

    frames_pad = videopadding(frames_t, s, T)
    masks_pad  = videopadding(masks_t,  s, T)

    recovered_frames = []

    # =====================
    # Inference
    # =====================
    with torch.no_grad():
        for j in range(n_frames):

            imgs = frames_pad[:, j : j + (T - 1) * s + 1 : s]
            msks = masks_pad[:,  j : j + (T - 1) * s + 1 : s]

            pred = net_G(imgs, msks)
            assert pred.dim() == 4, f"Unexpected output shape {pred.shape}"

            pred = (pred * 0.5 + 0.5).clamp(0, 1)

            frame_out = (
                pred.squeeze(0)
                .cpu()
                .numpy()
                .transpose(1, 2, 0)
            )

            recovered_frames.append(
                (frame_out * 255).astype(np.uint8)
            )

    # =====================
    # Write MP4
    # =====================
    out_path = os.path.join(out_dir, f"{out_video}.mp4")

    fourcc = cv2.VideoWriter_fourcc(*"mp4v")
    writer = cv2.VideoWriter(out_path, fourcc, FPS, (w, h))
    
    assert writer.isOpened(), "Failed to open VideoWriter"
    
    for frame in recovered_frames:
        # OpenCV expects BGR uint8
        frame_bgr = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)
        writer.write(frame_bgr)
    
    writer.release()
    
    print("Saved:", out_path)

    return out_path
```

```python
import os
import cv2

def extract_frame_png(video_path, frame_idx, out_dir, prefix=''):
    """
    Extract a single frame from an MP4 and save it as PNG.

    Args:
        video_path (str): path to input mp4
        frame_idx (int): 0-based frame index
        out_dir (str): directory to save png
        out_name (str, optional): output file name, defaults to video basename
    """

    os.makedirs(out_dir, exist_ok=True)

    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        raise RuntimeError(f"Cannot open video: {video_path}")

    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    if frame_idx < 0 or frame_idx >= total:
        cap.release()
        raise ValueError(
            f"Video has {total} frames, requested {frame_idx}"
        )

    # Seek to frame
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_idx)
    ret, frame = cap.read()
    cap.release()

    if not ret:
        raise RuntimeError(
            f"Failed to read frame {frame_idx} from {video_path}"
        )

    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    base = os.path.splitext(os.path.basename(video_path))[0]
    out_name = f"{prefix}_frame_{frame_idx}.png"

    out_path = os.path.join(out_dir, out_name)

    # cv2.imwrite expects BGR
    cv2.imwrite(out_path, cv2.cvtColor(frame_rgb, cv2.COLOR_RGB2BGR))

    return out_path
```

## Evaluation (video)

```python
import cv2
import numpy as np
from skimage.metrics import structural_similarity as ssim

def evaluate_video_quality(gt_video_path, pred_video_path, max_frames=None):
    """
    Args:
        gt_video_path   str, path to ground truth MP4
        pred_video_path str, path to predicted MP4
        max_frames      int or None, limit number of frames evaluated

    Returns:
        dict with mean MSE, PSNR, SSIM, DSSIM
        and per frame lists if needed
    """

    cap_gt = cv2.VideoCapture(gt_video_path)
    cap_pr = cv2.VideoCapture(pred_video_path)

    if not cap_gt.isOpened() or not cap_pr.isOpened():
        raise FileNotFoundError("Cannot open one of the videos")

    mse_list, psnr_list, ssim_list, dssim_list = [], [], [], []

    frame_count = 0

    while True:
        ret_gt, gt_bgr = cap_gt.read()
        ret_pr, pr_bgr = cap_pr.read()

        if not ret_gt or not ret_pr:
            break

        # Convert to RGB
        gt = cv2.cvtColor(gt_bgr, cv2.COLOR_BGR2RGB).astype(np.float32)
        pr = cv2.cvtColor(pr_bgr, cv2.COLOR_BGR2RGB).astype(np.float32)

        if gt.shape != pr.shape:
            raise ValueError("Frame size mismatch")

        # MSE
        mse = np.mean((gt - pr) ** 2)

        # PSNR
        psnr = float("inf") if mse == 0 else 20 * np.log10(255.0 / np.sqrt(mse))

        # SSIM
        ssim_val = ssim(
            gt,
            pr,
            channel_axis=2,
            data_range=255
        )

        dssim = (1 - ssim_val) / 2

        mse_list.append(mse)
        psnr_list.append(psnr)
        ssim_list.append(ssim_val)
        dssim_list.append(dssim)

        frame_count += 1
        if max_frames is not None and frame_count >= max_frames:
            break

    cap_gt.release()
    cap_pr.release()

    if frame_count == 0:
        raise RuntimeError("No frames evaluated")

    return {
        "MSE": np.mean(mse_list),
        "PSNR": np.mean(psnr_list),
        "SSIM": np.mean(ssim_list),
        "DSSIM": np.mean(dssim_list),
        "num_frames": frame_count,
        "per_frame": {
            "MSE": mse_list,
            "PSNR": psnr_list,
            "SSIM": ssim_list,
            "DSSIM": dssim_list,
        }
    }
```

```python
import cv2
import numpy as np

def evaluate_mask_video_quality(
    gt_mask_video_path,
    pred_mask_video_path,
    threshold=127,
    max_frames=None
):
    """
    Args:
        gt_mask_video_path   str, path to ground truth mask MP4
        pred_mask_video_path str, path to filtered / predicted mask MP4
        threshold             int, binarization threshold (0–255)
        max_frames            int or None, limit number of frames evaluated

    Returns:
        dict with mean IoU, Dice, Precision, Recall, PixelAcc
        and per frame lists
    """

    cap_gt = cv2.VideoCapture(gt_mask_video_path)
    cap_pr = cv2.VideoCapture(pred_mask_video_path)

    if not cap_gt.isOpened() or not cap_pr.isOpened():
        raise FileNotFoundError("Cannot open one of the mask videos")

    iou_list, dice_list = [], []
    prec_list, recall_list, acc_list = [], [], []

    frame_count = 0
    eps = 1e-8

    while True:
        ret_gt, gt_frame = cap_gt.read()
        ret_pr, pr_frame = cap_pr.read()

        if not ret_gt or not ret_pr:
            break

        # Convert to grayscale
        if gt_frame.ndim == 3:
            gt_gray = cv2.cvtColor(gt_frame, cv2.COLOR_BGR2GRAY)
        else:
            gt_gray = gt_frame

        if pr_frame.ndim == 3:
            pr_gray = cv2.cvtColor(pr_frame, cv2.COLOR_BGR2GRAY)
        else:
            pr_gray = pr_frame

        assert gt_gray.shape == pr_gray.shape, "Mask frame size mismatch"

        gt_bin = gt_gray > threshold
        pr_bin = pr_gray > threshold

        tp = np.logical_and(pr_bin, gt_bin).sum()
        fp = np.logical_and(pr_bin, ~gt_bin).sum()
        fn = np.logical_and(~pr_bin, gt_bin).sum()
        tn = np.logical_and(~pr_bin, ~gt_bin).sum()

        iou = tp / (tp + fp + fn + eps)
        dice = (2 * tp) / (2 * tp + fp + fn + eps)
        precision = tp / (tp + fp + eps)
        recall = tp / (tp + fn + eps)
        pixel_acc = (tp + tn) / (tp + tn + fp + fn + eps)

        iou_list.append(iou)
        dice_list.append(dice)
        prec_list.append(precision)
        recall_list.append(recall)
        acc_list.append(pixel_acc)

        frame_count += 1
        if max_frames is not None and frame_count >= max_frames:
            break

    cap_gt.release()
    cap_pr.release()

    if frame_count == 0:
        raise RuntimeError("No frames evaluated")

    return {
        "IoU": np.mean(iou_list),
        "Dice": np.mean(dice_list),
        "Precision": np.mean(prec_list),
        "Recall": np.mean(recall_list),
        "PixelAcc": np.mean(acc_list),
        "num_frames": frame_count,
        "per_frame": {
            "IoU": iou_list,
            "Dice": dice_list,
            "Precision": prec_list,
            "Recall": recall_list,
            "PixelAcc": acc_list,
        }
    }
```

## Evaluation (frame)

```python
import cv2
import numpy as np
from skimage.metrics import structural_similarity as ssim

def evaluate_frame_quality(gt_frame_path, pred_frame_path):
    """
    Args:
        gt_frame_path   str, path to ground truth PNG
        pred_frame_path str, path to predicted PNG

    Returns:
        dict with MSE, PSNR, SSIM, DSSIM
    """

    gt_bgr = cv2.imread(gt_frame_path)
    pred_bgr = cv2.imread(pred_frame_path)

    if gt_bgr is None or pred_bgr is None:
        raise FileNotFoundError("Failed to load one of the frames")

    # Convert to RGB
    gt = cv2.cvtColor(gt_bgr, cv2.COLOR_BGR2RGB).astype(np.float32)
    pred = cv2.cvtColor(pred_bgr, cv2.COLOR_BGR2RGB).astype(np.float32)

    # Ensure same size
    assert gt.shape == pred.shape, "Frame size mismatch"

    # MSE
    mse = np.mean((gt - pred) ** 2)

    # PSNR
    if mse == 0:
        psnr = float("inf")
    else:
        psnr = 20 * np.log10(255.0 / np.sqrt(mse))

    # SSIM
    ssim_val = ssim(
        gt,
        pred,
        channel_axis=2,
        data_range=255
    )

    # DSSIM
    dssim = (1 - ssim_val) / 2

    return {
        "MSE": mse,
        "PSNR": psnr,
        "SSIM": ssim_val,
        "DSSIM": dssim,
    }
```

```python
import cv2
import numpy as np

def evaluate_mask_quality(gt_mask_path, pred_mask_path, threshold=127):
    """
    Args:
        gt_mask_path   str, path to ground truth mask image
        pred_mask_path str, path to filtered / predicted mask image
        threshold      int, binarization threshold (0–255)

    Returns:
        dict with IoU, Dice, Precision, Recall, PixelAcc
    """

    gt = cv2.imread(gt_mask_path, cv2.IMREAD_GRAYSCALE)
    pred = cv2.imread(pred_mask_path, cv2.IMREAD_GRAYSCALE)

    if gt is None or pred is None:
        raise FileNotFoundError("Failed to load one of the mask images")

    assert gt.shape == pred.shape, "Mask size mismatch"

    gt_bin = gt > threshold
    pred_bin = pred > threshold

    tp = np.logical_and(pred_bin, gt_bin).sum()
    fp = np.logical_and(pred_bin, ~gt_bin).sum()
    fn = np.logical_and(~pred_bin, gt_bin).sum()
    tn = np.logical_and(~pred_bin, ~gt_bin).sum()

    eps = 1e-8

    iou = tp / (tp + fp + fn + eps)
    dice = (2 * tp) / (2 * tp + fp + fn + eps)
    precision = tp / (tp + fp + eps)
    recall = tp / (tp + fn + eps)
    pixel_acc = (tp + tn) / (tp + tn + fp + fn + eps)

    return {
        "IoU": iou,
        "Dice": dice,
        "Precision": precision,
        "Recall": recall,
        "PixelAcc": pixel_acc,
    }
```

```python
import os
import cv2

def extract_debug_frames_from_videos(results_dict, video_keys, frame_idx, out_dir="_final_debug_frames"):
    """
    Extract frame `frame_idx` from selected videos listed in `video_keys`,
    where videos come from the dictionary returned by full_process().

    Parameters
    ----------
    results_dict : dict
        Output from full_process(), containing filenames of generated videos.
    video_keys : list of str
        Keys in results_dict to extract frames from.
        Example: ["original_video", "noised_video", "denoised_video",
                  "original_mask", "noised_mask", "denoised_mask"]
    frame_idx : int
        Frame index to extract (0-indexed).
    out_dir : str
        Directory where frames will be saved.
    """
    os.makedirs(out_dir, exist_ok=True)

    for key in video_keys:
        if key not in results_dict:
            print(f"[WARN] Key '{key}' not found in results_dict. Skipping.")
            continue

        video_path = results_dict[key]
        video_path = os.path.expanduser(video_path)

        cap = cv2.VideoCapture(video_path)
        if not cap.isOpened():
            print(f"[ERROR] Cannot open video: {video_path}")
            continue

        total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        if frame_idx >= total:
            print(f"[WARN] Video '{key}' has only {total} frames. Requested {frame_idx}. Skipping.")
            cap.release()
            continue

        # Seek to frame
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_idx)
        ret, frame = cap.read()
        cap.release()

        if not ret:
            print(f"[ERROR] Failed to read frame {frame_idx} from {video_path}")
            continue

        # Convert to RGB for consistency
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        out_name = f"{key}_frame_{frame_idx}.png"
        out_path = os.path.join(out_dir, out_name)
        cv2.imwrite(out_path, cv2.cvtColor(frame_rgb, cv2.COLOR_RGB2BGR))

        print(f"[SAVED] {out_path}")

    print(f"[DONE] Extracted frames → {out_dir}")
```

```python
import numpy as np
import os
import cv2

def extract_debug_frames_from_videos(results, video_keys, frame_idx, out_dir):
    """
    Extracts a specific frame from multiple videos listed in `video_keys`,
    saves individual frames, and also saves a composite image.

    Parameters
    ----------
    results : dict
        The dict returned by full_process(), containing video paths.
    video_keys : list[str]
        Keys inside `results` whose videos will be sampled.
        Example: ["original_video", "noised_video", "original_mask", "noised_mask"]
    frame_idx : int
        Frame index to extract.
    out_dir : str
        Output directory where PNGs will be saved.

    Returns
    -------
    dict with list of saved paths.
    """

    os.makedirs(out_dir, exist_ok=True)

    # ---- helper to load a specific frame ----
    def load_frame(video_path, idx):
        cap = cv2.VideoCapture(video_path)
        if not cap.isOpened():
            raise FileNotFoundError(f"Cannot open video: {video_path}")

        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        ret, frame = cap.read()
        cap.release()

        if not ret:
            raise ValueError(f"Frame {idx} not found in {video_path}")

        return cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # -----------------------------------------
    # Load frames for all requested videos
    # -----------------------------------------
    frames = []
    saved_paths = []

    for key in video_keys:
        vpath = results[key]
        frame = load_frame(vpath, frame_idx)

        out_path = os.path.join(out_dir, f"{key}_frame_{frame_idx}.png")
        cv2.imwrite(out_path, cv2.cvtColor(frame, cv2.COLOR_RGB2BGR))

        saved_paths.append(out_path)
        frames.append(frame)

        print(f"[SAVED] {out_path}")

    # -----------------------------------------
    # Create composite image (flexible layout)
    # -----------------------------------------
    gap = np.full((frames[0].shape[0], 10, 3), 255, dtype=np.uint8)  # vertical gap

    # ---- 1 or 2 frames → horizontal strip ----
    if len(frames) <= 2:
        composite = frames[0]
        for i in range(1, len(frames)):
            composite = np.hstack([composite, gap, frames[i]])

    # ---- 3 frames → 1×3 row ----
    elif len(frames) == 3:
        composite = np.hstack([frames[0], gap, frames[1], gap, frames[2]])

    # ---- 4 frames → 2×2 grid ----
    else:
        g_h = np.zeros((10, frames[0].shape[1]*2 + 10, 3), dtype=np.uint8)  # horizontal gap

        top = np.hstack([frames[0], gap, frames[1]])
        bottom = np.hstack([frames[2], gap, frames[3]])
        composite = np.vstack([top, g_h, bottom])

    comp_path = os.path.join(out_dir, f"frame_{frame_idx}_composite.png")
    cv2.imwrite(comp_path, cv2.cvtColor(composite, cv2.COLOR_RGB2BGR))
    saved_paths.append(comp_path)

    print(f"[SAVED] {comp_path}")

    return {
        "individual_frames": saved_paths[:-1],
        "composite": comp_path
    }
```
