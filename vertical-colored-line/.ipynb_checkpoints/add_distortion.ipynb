```python
import cv2
import numpy as np
import random

def process_video_vertical_red_lines(input_path, output_path, 
                                     line_width=1, 
                                     min_stay_seconds=1.2):
    
    cap = cv2.VideoCapture(input_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    writer = cv2.VideoWriter(
        output_path,
        cv2.VideoWriter_fourcc(*"mp4v"),
        fps, (w, h)
    )

    
    min_frames = int(min_stay_seconds * fps)
    print(min_frames)
    
    x1 = -1
    frame_idx = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        
        if frame_idx % min_frames == 10:
            x1 = random.randint(int(w*0.1), int(0.9*w - 1))
            print(frame_idx, x1)
        
        out = frame.copy()
        cv2.line(out, (x1, 0), (x1, h-1), (0,255,255), line_width)

        writer.write(out)
        frame_idx += 1

    cap.release()
    writer.release()
    print("Done:", output_path)

```

```python
process_video_vertical_red_lines("./raw/test_1_raw.mp4","./distortion/test_1_distortion.mp4",)
```
